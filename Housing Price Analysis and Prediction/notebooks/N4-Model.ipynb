```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
x_train = pd.read_csv("../data/processed/x_train.csv")
x_val = pd.read_csv("../data/processed/x_val.csv")
```

```python
y_train = pd.read_csv("../data/processed/y_train.csv")
y_val = pd.read_csv("../data/processed/y_val.csv")
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
val_preds = model.predict(x_val)
val_rmse = np.sqrt(mean_squared_error(y_val, val_preds))
val_r2 = r2_score(y_val, val_preds)
```

```python
print("ðŸ“Œ Validation Results")
print("RMSE (Validation):", round(val_rmse, 3))
print("RÂ² (Validation):", round(val_r2, 3))
```

```python
plt.scatter(y_val, val_preds, alpha=0.5)
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Validation Predictions")
plt.show()
```

```python
x_test = pd.read_csv("../data/processed/x_test.csv")
y_test = pd.read_csv("../data/processed/y_test.csv")
```

```python
test_preds = model.predict(x_test)
test_rmse = np.sqrt(mean_squared_error(y_test, test_preds))
test_r2 = r2_score(y_test, test_preds)
```

```python
print("\nðŸ“Œ Test Results (Final)")
print("RMSE (Test):", round(test_rmse, 3))
print("RÂ² (Test):", round(test_r2, 3))
```

```python
plt.scatter(y_test, test_preds, color='blue')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r-')
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Predicted vs Actual')
plt.show()
```
